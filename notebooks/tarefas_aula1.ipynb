### Exercício C1 — Fatorial com `while`

O **fatorial** de um inteiro $n \geq 0$ — escrito $n!$ — é o produto de todos os inteiros positivos de 1 até $n$:

$$n! = n \times (n-1) \times (n-2) \times \cdots \times 2 \times 1$$

Por convenção, $0! = 1$. Por exemplo, $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$.

Calcule o fatorial de `n` usando um laço **`while`** — sem usar `math.factorial`.

**Casos de teste:**

| `n` | `n!` |
|---|---|
| 0  | 1 |
| 1  | 1 |
| 5  | 120 |
| 10 | 3628800 |

````python
n = 10

# TODO: calcular n! usando while
fatorial = ...

print(f'{n}! = {fatorial}')
````

````python
# @title 💡 Solução C1 (clique em "Mostrar código" para revelar)
# Solução C1
n = 10

fatorial = 1
i = 1
while i <= n:
    fatorial = fatorial * i
    i = i + 1

print(f'{n}! = {fatorial}')
````

### Exercício C2 — Conjectura de Collatz

Pegue um inteiro positivo `n` e aplique repetidamente a seguinte regra:

- Se `n` for **par**, divida por 2.
- Se `n` for **ímpar**, multiplique por 3 e some 1.

A **conjectura de Collatz** afirma que, para qualquer ponto de partida, eventualmente chegamos a 1 — esse é um dos problemas em aberto mais famosos da matemática, ainda não provado em geral.

Por exemplo, partindo de `n = 6` a sequência é:

```
6 → 3 → 10 → 5 → 16 → 8 → 4 → 2 → 1
```

Foram **8 passos** até chegar em 1.

Dado um `n` inicial, conte quantos passos a sequência leva até chegar em 1.

**Casos de teste:**

| `n` inicial | passos até 1 |
|---|---|
| 1  | 0 |
| 6  | 8 |
| 27 | 111 (famoso por demorar muito) |

````python
n = 27

# TODO: contar quantos passos a sequência de Collatz leva até chegar em 1
passos = 0
...

print(f'{passos} passos')
````

````python
# @title 💡 Solução C2 (clique em "Mostrar código" para revelar)
# Solução C2
n = 27

passos = 0
while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1
    passos = passos + 1

print(f'{passos} passos')
````

| `salario_anual` | `fracao_economizada` | `custo_da_casa` | meses |

````python
# @title 💡 Solução D3 (clique em "Mostrar código" para revelar)
# Solução D3
salario_anual = 112000
fracao_economizada = 0.17
custo_da_casa = 750000

fracao_entrada = 0.25
taxa_anual = 0.05

entrada = custo_da_casa * fracao_entrada
salario_mensal = salario_anual / 12
taxa_mensal = taxa_anual / 12

economias = 0
meses = 0
while economias < entrada:
    rendimento = economias * taxa_mensal
    economias = economias + salario_mensal * fracao_economizada + rendimento
    meses = meses + 1

print(f'Número de meses: {meses}')
````
