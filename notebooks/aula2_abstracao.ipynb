## Por que funções?

Suponha que a gente queira converter uma temperatura em Celsius para Fahrenheit. A fórmula é $F = C \cdot 9/5 + 32$.

````
c = 25
f = c * 9/5 + 32
f
````

Beleza, 25°C dá 77°F. Mas agora preciso converter mais uma:

````
c = 100
f = c * 9/5 + 32
f
````

E outra:

````
c = 0
f = c * 9/5 + 32
f
````

Toda vez que muda só o valor de entrada, a gente escreve a mesma fórmula de novo. Se a turma tivesse 30 alunos e cada um precisasse converter uma temperatura, seríamos 30 cópias da mesma linha — e qualquer mudança na fórmula teria que ser feita em 30 lugares.

É isso que uma **função** resolve. Uma função é um pedaço de código com nome, que pode ser **chamado** quantas vezes a gente quiser com entradas diferentes.

## Anatomia de uma função

A sintaxe básica:

```python
def nome_da_funcao(parametros):
    corpo
    return valor
```

Vamos começar com algo simples: uma função que devolve o maior entre dois números.

````
def max_val(x, y):
    if x > y:
        return x
    else:
        return y
````

Só de rodar a célula acima, a função `max_val` passa a existir. Pra usá-la, **invocamos** com argumentos:

````
max_val(3, 4)
````

````
max_val(10, 7)
````

Quatro coisas importantes acontecendo aqui:

1. `def` avisa o Python: "o que vem a seguir é a definição de uma função".
2. `x` e `y` são os **parâmetros formais** — nomes que valem só dentro do corpo da função.
3. Quando chamamos `max_val(3, 4)`, o `3` é amarrado ao `x` e o `4` ao `y`. Esses são os **argumentos** (ou parâmetros reais).
4. O `return` encerra a execução da função e devolve o valor.

Uma chamada de função é uma **expressão**: tem um valor. Por isso ela pode aparecer em qualquer lugar onde um valor pode aparecer.

````
max_val(3, 4) * max_val(5, 2)
````

````
dobro_do_maior = 2 * max_val(8, 11)
dobro_do_maior
````

Vamos resolver o nosso problema da temperatura. Em vez de copiar e colar a fórmula a cada conversão, escrevemos uma função:

````
def celsius_para_fahrenheit(c):
    return c * 9/5 + 32
````

````
celsius_para_fahrenheit(25)
````

````
celsius_para_fahrenheit(100)
````

````
celsius_para_fahrenheit(0)
````

Uma única função, vários usos. Se um dia a fórmula mudar (ou a gente encontrar um *bug*), ajustamos em **um único lugar** — todos os chamadores se beneficiam automaticamente.

Outro exemplo, agora com dois parâmetros: a área de um retângulo.

````
def area_retangulo(base, altura):
    return base * altura
````

````
area_retangulo(3, 4)
````

````
area_retangulo(10, 5)
````

Isso é o que se chama **abstração**: a gente esconde o **como** atrás de um nome (`area_retangulo`) e passa a usar só o **o quê**. Quem chama a função não precisa lembrar a fórmula.

## Argumentos posicionais, nomeados e com valor padrão

Até aqui passamos argumentos pela **posição**: o primeiro vai pro primeiro parâmetro, o segundo pro segundo, e assim por diante. Mas Python aceita outras duas formas: argumentos **nomeados** (também chamados *keyword arguments*) e **valores padrão**.

````
def imprimir_nome(primeiro, sobrenome, invertido):
    if invertido:
        print(sobrenome + ', ' + primeiro)
    else:
        print(primeiro, sobrenome)
````

````
imprimir_nome('Carlos', 'Melo', False)
````

````
imprimir_nome('Carlos', 'Melo', invertido=True)
````

Os argumentos nomeados deixam a chamada mais legível: olhando `invertido=True`, sabemos exatamente o que aquele `True` significa.

Argumentos com **valor padrão** vão ainda mais longe: a gente declara um valor que será usado caso o chamador não passe nada.

````
def imprimir_nome(primeiro, sobrenome, invertido=False):
    if invertido:
        print(sobrenome + ', ' + primeiro)
    else:
        print(primeiro, sobrenome)
````

````
imprimir_nome('Carlos', 'Melo')
````

````
imprimir_nome('Carlos', 'Melo', invertido=True)
````

Note como a função agora pode ser chamada de duas formas, dependendo do que o usuário quer. **Valores padrão tornam a função mais flexível sem complicar o uso comum**.

## Escopo: onde cada variável vive

Considere este código. Antes de rodar, tente prever o que ele imprime.

````
def f(x):
    y = 1
    x = x + y
    print('dentro de f: x =', x)
    return x
````

````
x = 3
y = 2
z = f(x)
````

````
print('x =', x)
print('y =', y)
print('z =', z)
````

Olha o que acontece: dentro da função, `x` virou `4`. Mas **fora** da função, `x` continua sendo `3`. Por quê?

Cada função tem o seu próprio **escopo** (também chamado *namespace*). O parâmetro `x` dentro de `f` é uma variável **diferente** do `x` que existe lá fora — ela só vive durante a execução de `f` e some quando a função retorna. Reatribuir `x` dentro da função não toca no `x` de fora.

Você pode pensar assim: existe uma **tabela de variáveis** no nível principal do programa, e cada chamada de função cria uma tabela **nova**, que vive enquanto a função estiver executando e some quando ela retorna. É isso que permite duas funções usarem o nome `x` sem se atrapalharem.

````
def soma_um(n):
    n = n + 1   # mexe no n local; o n de fora não se altera
    return n

n = 10
resultado = soma_um(n)
print('n fora:', n)
print('resultado:', resultado)
````

**Regra prática**: se uma função precisa de um valor, ela recebe como parâmetro. Se ela produz um valor, ela devolve com `return`. Funções que só entram-e-saem por essas duas vias são fáceis de entender, de testar e de reaproveitar.

## Especificação: o contrato da função

Uma função bem escrita tem uma **especificação** — uma descrição curta do que ela faz, do que ela espera receber, e do que ela devolve. Em Python isso vai numa *docstring*, entre aspas triplas, logo depois do `def`.

````
def area_retangulo(base, altura):
    """Calcula a área de um retângulo.

    base e altura devem ser números positivos.
    Retorna base * altura.
    """
    return base * altura
````

````
help(area_retangulo)
````

A docstring é o **contrato** da função: ela diz pra quem usa o que esperar, sem precisar ler o corpo. Essa separação entre o **o quê** (especificação) e o **como** (implementação) é o coração da abstração.

## Mutabilidade e *aliasing* — o destaque da aula

Em Python, alguns objetos são **mutáveis** (podem ser alterados depois de criados) e outros são **imutáveis** (uma vez criados, não mudam). Listas são mutáveis. Inteiros, *floats*, *booleans*, *strings* e tuplas são imutáveis.

Tente alterar o primeiro caractere de uma string:

````
s = 'Sigmoidal'
try:
    s[0] = 'X'
except TypeError as e:
    print('Erro:', e)
````

Agora a mesma coisa com lista:

````
lst = [1, 2, 3]
lst[0] = 99
lst
````

Funciona. E é exatamente essa flexibilidade que nos coloca em apuros. Considere o seguinte cenário: a gente tem dois grupos de universidades.

````
militares = ['ITA', 'IME']
estaduais = ['USP', 'Unicamp', 'Unesp']
````

E queremos juntar as duas categorias num único objeto. Tem dois jeitos que **parecem** equivalentes:

````
universidades = [militares, estaduais]
universidades1 = [['ITA', 'IME'], ['USP', 'Unicamp', 'Unesp']]

print('universidades :', universidades)
print('universidades1:', universidades1)
````

````
universidades == universidades1
````

São iguais em **valor**. Mas serão o mesmo objeto? Existe um operador em Python pra perguntar isso: `is`.

````
universidades is universidades1
````

````
universidades[0] is universidades1[0]
````

`==` pergunta "esses dois têm o mesmo conteúdo?". `is` pergunta "esses dois são o mesmo objeto?". Olha o que acontece quando a gente altera a lista `militares` lá em cima:

````
militares.append('AMAN')
print('militares     :', militares)
print('universidades :', universidades)
print('universidades1:', universidades1)
````

**Universidades** mudou. **Universidades1** não. Isso porque `universidades` guarda uma **referência** à mesma lista `militares` — não uma cópia. Quando `militares` muda, `universidades[0]` muda junto, porque eles são **o mesmo objeto** com nomes diferentes. A esse fenômeno damos o nome de ***aliasing***: dois nomes apontando pro mesmo objeto.

O mesmo problema aparece numa forma ainda mais sutil. Suponha que eu queira fazer uma cópia de uma lista:

````
lista_a = [1, 2, 3]
lista_b = lista_a       # parece uma cópia
lista_b.append(99)
print('lista_a:', lista_a)
print('lista_b:', lista_b)
````

**Lista_a também ganhou o 99.** Porque `lista_b = lista_a` não copia: só dá um segundo nome ao mesmo objeto. Pra fazer uma cópia de verdade, a forma mais simples é com *slicing*:

````
lista_a = [1, 2, 3]
lista_b = lista_a[:]    # cópia rasa de verdade
lista_b.append(99)
print('lista_a:', lista_a)
print('lista_b:', lista_b)
````

O `[:]` cria uma lista **nova** com os mesmos elementos da original. (Existem também `list(lista_a)` e `lista_a.copy()`, que fazem a mesma coisa.)

Onde isso mais machuca? Quando passamos uma lista pra uma função. Olha:

````
def zerar_negativos(lista):
    for i in range(len(lista)):
        if lista[i] < 0:
            lista[i] = 0
    return lista

valores = [3, -1, 4, -2, 5]
saida = zerar_negativos(valores)
print('saida   :', saida)
print('valores :', valores)   # !!!
````

A lista `valores` foi **modificada silenciosamente**. A função recebeu uma referência à mesma lista que existe lá fora — e ao mexer nessa referência, mexeu no objeto original. Esse é o tipo de *bug* que custa horas: o programa imprime um resultado correto e, mais adiante, o estado de outra variável está estranho sem motivo aparente.

**Conclusão prática:** quando passar uma lista pra uma função, decida conscientemente: ela deve modificar a lista original, ou deve devolver uma versão nova? Se for o segundo caso, faça uma cópia logo no começo:

````
def zerar_negativos_seguro(lista):
    saida = lista[:]
    for i in range(len(saida)):
        if saida[i] < 0:
            saida[i] = 0
    return saida

valores = [3, -1, 4, -2, 5]
saida = zerar_negativos_seguro(valores)
print('saida   :', saida)
print('valores :', valores)
````

Esse é, sem exagero, o conceito mais importante da aula de hoje. Em Python científico — *numpy*, *pandas*, *pytorch* — quase todos os *bugs* sutis que vocês vão encontrar passam por aqui.

## Tuplas: quando você quer imutabilidade

Uma tupla é uma sequência **imutável**. A sintaxe usa parênteses (ou simplesmente vírgulas).

````
ponto = (3, 4)
ponto
````

````
ponto[0]
````

````
try:
    ponto[0] = 99
except TypeError as e:
    print('Erro:', e)
````

Por que usar tupla se a lista é mais flexível? Três motivos:

1. **Intenção**: se o valor não deve mudar, usar tupla deixa isso explícito.
2. **Segurança**: ninguém vai modificar acidentalmente.
3. **Múltiplos retornos**: quando uma função precisa devolver mais de um valor, a forma natural é uma tupla.

Os dois usos mais comuns: **desempacotamento** e **múltiplos retornos**.

````
x, y = ponto
print('x =', x)
print('y =', y)
````

````
def min_e_max(lista):
    return min(lista), max(lista)
````

````
menor, maior = min_e_max(notas)
print('menor:', menor)
print('maior:', maior)
````

**Exemplo prático:** contar a frequência de letras numa palavra.

````
palavra = 'visao_computacional'
frequencia = {}

for letra in palavra:
    if letra in frequencia:
        frequencia[letra] = frequencia[letra] + 1
    else:
        frequencia[letra] = 1

frequencia
````

## Recursão: funções que chamam a si mesmas

Algumas estruturas e processos são **recursivos** por natureza: a definição de uma pasta dentro de pasta dentro de pasta, a estrutura de um documento HTML, a definição matemática do fatorial. Pra esses, faz sentido escrever a solução de forma recursiva.

Uma função recursiva tem **duas partes**:

- **Caso base**: a entrada mais simples, cuja resposta a gente sabe sem chamar mais ninguém.
- **Passo recursivo**: a entrada geral, reescrita em termos de uma entrada menor.

Começamos pelo fatorial, na versão **iterativa** (com `for`) que você já saberia escrever:

````
def fatorial_iterativo(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado = resultado * i
    return resultado

fatorial_iterativo(5)
````

Agora a **versão recursiva**. A definição matemática é: $n! = n \cdot (n-1)!$ se $n > 1$, e $1! = 1$. Traduzindo direto:

````
def fatorial(n):
    if n <= 1:                     # caso base
        return 1
    return n * fatorial(n - 1)     # passo recursivo

fatorial(5)
````

````
fatorial(10)
````

Pra ter ideia do que está acontecendo, imagine `fatorial(4)`:

```
fatorial(4) = 4 * fatorial(3)
            = 4 * (3 * fatorial(2))
            = 4 * (3 * (2 * fatorial(1)))
            = 4 * (3 * (2 * 1))
            = 24
```

Cada chamada empilha uma nova execução; a pilha desempila quando o caso base é atingido.

Outro exemplo clássico: a sequência de **Fibonacci**, definida por $F(0) = 0$, $F(1) = 1$, e $F(n) = F(n-1) + F(n-2)$ para $n \geq 2$.

````
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

[fib(i) for i in range(10)]
````

Bonito e direto. Mas tem uma armadilha: experimente calcular `fib(33)`.

````
import time

inicio = time.time()
resultado = fib(33)
tempo = time.time() - inicio
print('fib(33) =', resultado, '   tempo:', round(tempo, 3), 's')
````

Pra um número tão pequeno, a função demorou uma fração visível de segundo. O motivo é que `fib(n)` recalcula a **mesma** subexpressão muitas vezes — `fib(31)` é calculado por dois ramos diferentes, `fib(30)` por três, e assim por diante. O número de chamadas cresce **exponencialmente** com `n`.

Esse é o gancho perfeito pra Aula 4, quando a gente vai falar sobre **complexidade**. Por enquanto, fica a observação: o fato de uma função ser fácil de escrever recursivamente não significa que ela seja eficiente.

**Quando usar recursão na prática?** Quando a estrutura do problema é naturalmente recursiva (árvores, partições, divisão e conquista) e o número de chamadas é razoável. Pra coisas como Fibonacci puro, é melhor o `for`.
