```python
def prefixo_comum(palavras):
    # TODO: comparar as palavras posição por posição
    ...

prefixo_comum(['flor', 'flores', 'flutuar'])
```

```python
# @title 💡 Solução A4 (clique em "Mostrar código" para revelar)
def prefixo_comum(palavras):
    prefixo = ''
    mais_curta = min(palavras, key=len)
    for i in range(len(mais_curta)):
        caractere = palavras[0][i]
        for palavra in palavras:
            if palavra[i] != caractere:
                return prefixo
        prefixo += caractere
    return prefixo

assert prefixo_comum(['flor', 'flores', 'flutuar']) == 'fl'
assert prefixo_comum(['cachorro', 'gato', 'rato']) == ''
assert prefixo_comum(['casa', 'casamento', 'casaco']) == 'casa'
print('OK!')
```

```python
# @title 💡 Solução B3 (clique em "Mostrar código" para revelar)
def parenteses_validos(s):
    pares = {')': '(', ']': '[', '}': '{'}
    pilha = []
    for c in s:
        if c in '([{':
            pilha.append(c)
        else:
            if len(pilha) == 0 or pilha.pop() != pares[c]:
                return False
    return len(pilha) == 0

assert parenteses_validos('()') == True
assert parenteses_validos('()[]{}') == True
assert parenteses_validos('(]') == False
assert parenteses_validos('([)]') == False
assert parenteses_validos('{[]}') == True
print('OK!')
```

```python
# @title 💡 Solução C4 (clique em "Mostrar código" para revelar)
def busca_sequencial(nums, alvo):
    for i in range(len(nums)):
        if nums[i] == alvo:
            return i
    return -1

def busca_binaria(nums, alvo):
    baixo, alto = 0, len(nums) - 1
    while baixo <= alto:
        meio = (baixo + alto) // 2
        if nums[meio] == alvo:
            return meio
        if nums[meio] < alvo:
            baixo = meio + 1
        else:
            alto = meio - 1
    return -1

for funcao in [busca_sequencial, busca_binaria]:
    assert funcao([-1, 0, 3, 5, 9, 12], 9) == 4
    assert funcao([-1, 0, 3, 5, 9, 12], 2) == -1
    assert funcao([5], 5) == 0
print('OK!')
```
