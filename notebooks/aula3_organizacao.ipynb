## Por que separar código em arquivos

Até aqui, assumimos que o programa inteiro cabe em um único arquivo. Isso é razoável enquanto ele é pequeno. Em projetos reais, porém, com várias pessoas contribuindo simultaneamente, manter tudo num arquivo só vira um problema imediato: conflitos a cada edição, dificuldade de localizar partes específicas, código praticamente impossível de reaproveitar em outro projeto.

A solução do Python é o **módulo**: um arquivo `.py` contendo definições (funções, variáveis, classes) que outros programas podem importar e utilizar. Vamos construir um.

A célula abaixo usa um comando especial do Jupyter/Colab, `%%writefile`, que grava o conteúdo da célula em um arquivo. Estamos criando o módulo `circulo.py`.

```python
%%writefile circulo.py
pi = 3.14159

def area(raio):
    return pi * (raio**2)

def circunferencia(raio):
    return 2 * pi * raio

def superficie_esfera(raio):
    return 4.0 * area(raio)
```

Com o arquivo criado, qualquer programa pode acessá-lo por meio de `import`. O nome do módulo é o nome do arquivo, sem o `.py`.

```python
import circulo
```

Para usar o que está definido no módulo, utilizamos a **notação de ponto**: `modulo.nome`.

```python
circulo.pi
```

```python
circulo.area(3)
```

```python
circulo.circunferencia(3)
```

```python
circulo.superficie_esfera(3)
```

Por que o prefixo `circulo.` em tudo? Porque cada módulo mantém o seu **próprio espaço de nomes** (em inglês, *namespace*). Isso evita colisões: um `pi` definido no programa principal é independente do `pi` que vive dentro do módulo.

```python
pi = 3
print(pi)
print(circulo.pi)
```

O `pi` local passou a valer `3`, enquanto `circulo.pi` continua valendo `3.14159`. São objetos distintos, em espaços de nomes distintos, e é a notação de ponto que impede que um sobrescreva o outro.

Existe uma variante, `from modulo import *`, que traz todos os nomes do módulo para o espaço atual. Assim, podemos chamar `area` diretamente, sem o prefixo `circulo.`.

```python
from circulo import *
area(3)
```

Funciona, mas o `import *` é considerado má prática. Ao ler `area(3)` em um programa maior, fica difícil rastrear de qual módulo a função veio. Como diz o *Zen of Python*, **explícito é melhor que implícito**.

A forma mais comum no dia a dia é `import modulo as apelido`, que dá um nome mais curto a um módulo. Você verá esse padrão repetidas vezes: `import numpy as np`, `import pandas as pd`. Por enquanto, importemos um módulo da biblioteca padrão do Python: o `math`.

```python
import math
```

```python
math.log(8, 2)
```

```python
math.sqrt(2)
```

O módulo `math` traz dezenas de funções matemáticas e constantes como `math.pi`. Essa é a grande vantagem dos módulos: reaproveitar código já escrito e testado por outras pessoas, em vez de reimplementar do zero.

1. importa-se **um módulo por linha**;
2. os imports ficam no **início do arquivo**;

## Arquivos: persistência entre execuções

Todo programa útil precisa, em algum momento, **ler** dados externos ou **gravar** resultados para uso posterior. Quando o programa termina, tudo que estava na memória é descartado, a menos que tenha sido salvo em um arquivo.

O Python acessa arquivos por meio de um *file handle*, obtido pela função `open`. Vamos criar um arquivo e gravar alguns nomes nele.

```python
arquivo = open('alunos.txt', 'w')
for nome in ['Ana', 'Beto', 'Cris']:
    arquivo.write(nome + '\n')
arquivo.close()
```

Três aspectos a comentar. O modo `'w'` (de *write*) abre o arquivo **para escrita**. O método `write` grava texto, e o `'\n'` é o caractere de **quebra de linha**, que separa um nome do outro. Por fim, `close()` fecha o arquivo. **Sempre feche** o que abrir; caso contrário, parte do conteúdo escrito pode não chegar ao disco.

Agora leremos o arquivo de volta. O modo `'r'` (de *read*) abre para leitura. Como o Python trata um arquivo como uma sequência de linhas, é possível percorrê-lo com um `for`.

```python
arquivo = open('alunos.txt', 'r')
for linha in arquivo:
    print(linha)
arquivo.close()
```

Apareceu uma linha em branco entre os nomes. Isso ocorre porque cada linha do arquivo já termina com `\n` e o `print` acrescenta uma quebra adicional. Para evitar, removemos o último caractere com `linha[:-1]`.

```python
arquivo = open('alunos.txt', 'r')
for linha in arquivo:
    print(linha[:-1])
arquivo.close()
```

Note que precisamos chamar `close()` manualmente todas as vezes. Existe uma forma mais segura: o bloco `with`. Ele abre o arquivo, disponibiliza-o dentro do bloco e **fecha automaticamente** ao final, inclusive em caso de erro no meio.

```python
with open('alunos.txt', 'r') as arquivo:
    for linha in arquivo:
        print(linha[:-1])
```

Daqui em diante, esta é a forma recomendada de manipular arquivos.

Resta um detalhe importante: os **modos** de abertura. O `'w'` **apaga** o conteúdo existente e escreve do zero. Para **acrescentar** ao final sem perder o que já estava lá, usa-se `'a'` (de *append*).

```python
with open('alunos.txt', 'a') as arquivo:
    arquivo.write('Davi\n')
```

```python
with open('alunos.txt') as arquivo:
    print(arquivo.read())
```

Resumindo os três modos mais frequentes:

- `'r'`: ler (é o padrão; o arquivo precisa existir);
- `'w'`: escrever do zero (apaga o conteúdo anterior);
- `'a'`: acrescentar ao final.

## Exceções: quando o programa encontra o inesperado

Apesar do nome, **exceções** não são raras em Python. Estão por toda parte, e você provavelmente já encontrou várias. Observe o que ocorre ao tentar acessar uma posição inexistente em uma lista:

```python
teste = [1, 2, 3]
teste[3]
```

O Python levantou um `IndexError` e interrompeu a execução. Quando uma exceção não é tratada, o programa **termina abruptamente** e exibe essa mensagem (o *traceback*). Os tipos mais comuns que aparecem na prática:

- `IndexError`: índice fora do alcance de uma sequência;
- `TypeError`: operação entre tipos incompatíveis (por exemplo, `1 + 'a'`);
- `NameError`: uso de um nome que não existe;
- `ValueError`: valor de tipo correto, mas inadequado (por exemplo, `int('abc')`);
- `ZeroDivisionError`: divisão por zero.

Uma exceção, contudo, **não precisa** derrubar o programa. Em muitos casos, é possível antecipar que algo pode dar errado e tratar a situação de forma controlada, usando o bloco `try`/`except`.

Considere o cálculo da média de gols por partida de um time: basta dividir o total de gols pelo número de partidas. Quase sempre funciona. Mas e se o time ainda não tiver entrado em campo?

```python
gols = 8
partidas = 0

try:
    media = gols / partidas
    print('Média de gols por partida:', media)
except ZeroDivisionError:
    print('O time ainda não disputou nenhuma partida.')
```

O Python tenta executar o bloco `try`. Se tudo correr bem, o `except` é ignorado. Caso uma `ZeroDivisionError` ocorra no meio do caminho (dividir por zero partidas), o controle passa imediatamente para o bloco `except`, e o programa segue adiante sem interromper.

É possível capturar a mensagem da exceção atribuindo-a a um nome com `as`. Vejamos a conversão de um texto que não representa um número:

```python
try:
    val = int('abc')
except ValueError as e:
    print('Deu ValueError:', e)
```

Um caso de uso clássico é a validação de entradas. Considere uma função que recebe um texto e devolve o inteiro correspondente. Se a entrada for inválida, em vez de interromper a execução, a função emite um aviso e devolve `None`.

```python
def para_int(texto):
    try:
        return int(texto)
    except ValueError:
        print(texto, 'não é um inteiro')
        return None
```

```python
para_int('42')
```

```python
para_int('abc')
```

Até aqui apenas **tratamos** exceções levantadas pelo próprio Python. Mas o programador também pode **levantar** exceções suas, com a instrução `raise`. Isso é útil quando uma função não consegue cumprir a sua especificação.

Como exemplo, uma função que devolve o primeiro número par de uma lista e sinaliza erro se não houver nenhum:

```python
def primeiro_par(L):
    for x in L:
        if x % 2 == 0:
            return x
    raise ValueError('a lista não contém nenhum par')
```

```python
primeiro_par([3, 7, 4, 9])
```

```python
try:
    primeiro_par([1, 3, 5])
except ValueError as msg:
    print('Capturado:', msg)
```

Por fim, o `assert`. Trata-se de uma forma direta de **programação defensiva**: declara-se uma condição que deve ser verdadeira naquele ponto da execução. Se for, o programa prossegue; se não for, um `AssertionError` é levantado imediatamente, com a mensagem fornecida.

```python
def media_das_notas(notas):
    assert len(notas) > 0, 'a lista de notas está vazia'
    return sum(notas) / len(notas)
```

```python
media_das_notas([7.5, 8.0, 9.5])
```

```python
try:
    media_das_notas([])
except AssertionError as e:
    print('AssertionError:', e)
```

O `assert` faz o erro aparecer **cedo e no ponto exato**, em vez de permitir que o programa continue com um dado inválido e falhe mais adiante, em um lugar onde a causa fica difícil de rastrear.

## Classes: dados e comportamento juntos

Chegamos ao ponto alto da aula. Até aqui, usamos apenas os tipos embutidos do Python: `int`, `str`, `list`, `dict`. Mas e quando o problema pede um tipo que o Python não oferece?

A ideia central da **orientação a objetos** é tratar objetos como pacotes que reúnem, em um único lugar, **os dados** e **as operações** que atuam sobre eles. Definimos novos tipos com a palavra `class`.

Vamos construir, passo a passo, um tipo "conjunto de inteiros": uma coleção de números sem repetições.

Começamos pelo esqueleto: o método especial `__init__`, executado a cada criação de objeto e responsável por estabelecer o estado inicial.

```python
class ConjuntoDeInteiros:
    def __init__(self):
        self._vals = []
    def insere(self, e):
        if e not in self._vals:
            self._vals.append(e)
```

Duas peças novas. O método `__init__` cria a lista `_vals`, que guardará os números; essa lista é um **atributo** do objeto. O `self`, presente como primeiro parâmetro de todo método, é o **próprio objeto**: por meio dele cada método acessa os dados da sua instância.

Para criar um objeto (uma *instância*) da classe, chamamos o nome dela como se fosse uma função. Em seguida, os métodos são acessados pela mesma notação de ponto dos módulos.

```python
s = ConjuntoDeInteiros()
s.insere(3)
s.insere(4)
s.insere(3)
s._vals
```

Inserimos o `3` duas vezes, mas ele aparece apenas uma. A lógica de "não repetir" está encapsulada no método `insere`, e o código que usa o conjunto não precisa saber como ela é implementada.

Note que `s.insere(3)` é chamado com **um** argumento, embora o método declare **dois** parâmetros (`self` e `e`). O objeto antes do ponto (`s`) é passado automaticamente como `self`. Vamos enriquecer a classe com mais métodos.

```python
class ConjuntoDeInteiros:
    def __init__(self):
        self._vals = []
    def insere(self, e):
        if e not in self._vals:
            self._vals.append(e)
    def contem(self, e):
        return e in self._vals
    def remove(self, e):
        try:
            self._vals.remove(e)
        except ValueError:
            raise ValueError(str(e) + ' não está no conjunto')
```

```python
s = ConjuntoDeInteiros()
s.insere(3)
print(s.contem(3))
print(s.contem(7))
```

Note como os tópicos da aula se conectam: o método `remove` usa `try`/`except` internamente e ainda **levanta** um `ValueError` mais informativo quando o elemento não está presente.

Há um incômodo, contudo: ao tentar imprimir o conjunto, o Python não sabe como exibi-lo de forma legível.

```python
print(s)
```

Aquele `<__main__.ConjuntoDeInteiros object at 0x...>` é a representação genérica que o Python utiliza quando não há uma definição específica de como converter o objeto em texto. A solução é definir o método especial `__str__`.

Métodos com dois sublinhados no começo e no fim, como `__init__` e `__str__`, são os **dunder methods** (de *double underscore*), também conhecidos como *magic methods*. O Python invoca o `__str__` automaticamente sempre que precisa converter o objeto em texto, por exemplo dentro de um `print`.

```python
class ConjuntoDeInteiros:
    def __init__(self):
        self._vals = []
    def insere(self, e):
        if e not in self._vals:
            self._vals.append(e)
    def __str__(self):
        self._vals.sort()
        return '{' + ','.join(str(e) for e in self._vals) + '}'
```

```python
s = ConjuntoDeInteiros()
s.insere(4)
s.insere(3)
print(s)
```

Agora `print(s)` exibe `{3,4}`, como se esperaria de um conjunto. O objeto sabe descrever a si mesmo.

Os dunder methods vão muito além do `__str__`. Eles permitem que os nossos tipos se comportem como os tipos nativos, inclusive em relação a **operadores**. Alguns dos principais:

- `__add__` para o operador `+`;
- `__eq__` para o operador `==`;
- `__lt__` para o operador `<`;
- `__len__` para a função `len`.

Vamos definir um vetor 2D que sabe somar e comparar.

```python
class Vetor2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return '(' + str(self.x) + ', ' + str(self.y) + ')'
    def __add__(self, outro):
        return Vetor2D(self.x + outro.x, self.y + outro.y)
    def __eq__(self, outro):
        return self.x == outro.x and self.y == outro.y
```

```python
a = Vetor2D(1, 2)
b = Vetor2D(3, 4)
print(a + b)
```

```python
a == Vetor2D(1, 2)
```

Ao escrever `a + b`, o Python efetivamente chama `a.__add__(b)`. Com esses métodos definidos, nossos objetos passam a usar a mesma sintaxe natural dos números e das listas.

## Um exemplo mais próximo do nosso domínio: Pessoa

Considere um sistema para gerenciar alunos e professores de um programa de pós-graduação. Antes de criar listas e dicionários soltos, vale identificar uma **abstração** comum: todos eles são, em última instância, pessoas com um nome. Vamos modelar essa ideia.

```python
class Pessoa:
    def __init__(self, nome):
        self._nome = nome
        try:
            self._sobrenome = nome[nome.rindex(' ') + 1:]
        except ValueError:
            self._sobrenome = nome
    def get_nome(self):
        return self._nome
    def get_sobrenome(self):
        return self._sobrenome
    def __lt__(self, outra):
        return self._nome < outra._nome
    def __str__(self):
        return self._nome
```

O `__init__` extrai o sobrenome a partir do último espaço do nome. Se não houver espaço algum, `rindex` levanta um `ValueError`, tratado assumindo-se que o nome completo é o próprio sobrenome. Novamente, uma exceção surgindo de forma natural dentro de uma classe.

```python
eu = Pessoa('Carlos Melo')
ele = Pessoa('Machado de Assis')
ela = Pessoa('Anitta')
print(ele.get_sobrenome())
print(ela.get_sobrenome())
```

Como definimos um `__lt__` em `Pessoa` (o operador `<`, comparando pelo nome), ganhamos um efeito útil: qualquer ferramenta do Python que dependa de comparações passa a funcionar com objetos `Pessoa`. O `sort` de uma lista é o exemplo mais direto.

```python
turma = [eu, ele, ela]
turma.sort()
for p in turma:
    print(p)
```

A lista foi ordenada alfabeticamente sem que precisássemos escrever nenhuma lógica de ordenação: o `sort` utilizou o nosso `__lt__` internamente.

## Herança: aproveitando o que já existe

Um aluno **é** uma pessoa (tem nome, sobrenome, tudo o que a `Pessoa` tem), mas possui também atributos próprios, como um número de matrícula. Reescrever tudo do zero seria desperdício. A **herança** permite criar um tipo novo que aproveita (herda) todo o conteúdo de um tipo existente e acrescenta o que for específico.

```python
class Aluno(Pessoa):
    _proxima_matricula = 0
    def __init__(self, nome):
        super().__init__(nome)
        self._matricula = Aluno._proxima_matricula
        Aluno._proxima_matricula += 1
    def get_matricula(self):
        return self._matricula
```

`class Aluno(Pessoa)` declara que `Aluno` é uma **subclasse** de `Pessoa`. O `super().__init__(nome)` chama o `__init__` da superclasse, reaproveitando a lógica do nome. Note também o `_proxima_matricula`: trata-se de uma **variável de classe** (pertence à classe, não a cada instância), e é o que permite atribuir uma matrícula única e crescente a cada novo aluno.

```python
a1 = Aluno('Barbara Beaver')
a2 = Aluno('João Silva')
print(a1, 'tem matrícula', a1.get_matricula())
print(a2, 'tem matrícula', a2.get_matricula())
```

O `a1` herdou o `__str__` de `Pessoa` (por isso `print(a1)` exibe o nome) e adquiriu o método `get_matricula`, que existe apenas em `Aluno`.

Uma subclasse também pode **substituir** (sobrescrever) um método herdado. Faz sentido que dois alunos sejam ordenados pela matrícula, e não pelo nome. Basta definir um novo `__lt__` em `Aluno`.

```python
class Aluno(Pessoa):
    _proxima_matricula = 0
    def __init__(self, nome):
        super().__init__(nome)
        self._matricula = Aluno._proxima_matricula
        Aluno._proxima_matricula += 1
    def get_matricula(self):
        return self._matricula
    def __lt__(self, outro):
        return self._matricula < outro._matricula
```

```python
a1 = Aluno('Barbara Beaver')
a2 = Aluno('João Silva')
print('a1 < a2 =', a1 < a2)
```

Agora `a1 < a2` compara matrículas. Quando um método é sobrescrito, prevalece a versão associada ao tipo do objeto utilizado na chamada: objetos do tipo `Aluno` usam o `__lt__` de `Aluno`; objetos do tipo `Pessoa` continuam utilizando o de `Pessoa`.

A herança pode ter múltiplos níveis. No contexto da pós-graduação, um aluno pode ser mestrando ou doutorando. Podemos criar uma camada intermediária `Estudante` e, abaixo dela, os dois tipos específicos.

```python
class Estudante(Aluno):
    pass

class Mestrando(Estudante):
    pass

class Doutorando(Estudante):
    def __init__(self, nome, orientador=None):
        super().__init__(nome)
        self._orientador = orientador
    def get_orientador(self):
        return self._orientador
```

O `pass` indica que a classe não acrescenta nada de novo: ela existe apenas para introduzir um tipo a mais na hierarquia. O `Doutorando`, por sua vez, recebe um orientador.

```python
m = Mestrando('Buzz Aldrin')
d = Doutorando('Neil Armstrong', orientador='Carlos Melo')
print(d.get_orientador())
```

Para verificar o tipo de um objeto, a ferramenta apropriada é a função `isinstance`. Diferentemente da comparação direta com `type`, ela respeita a hierarquia de herança.

```python
print(isinstance(m, Estudante))
print(isinstance(d, Aluno))
print(isinstance(d, Pessoa))
```

O objeto `d` é um `Doutorando`, mas também é, simultaneamente, um `Estudante`, um `Aluno` e uma `Pessoa`, porque cada um herda do anterior. O `isinstance` reconhece toda essa linhagem.

Uma última convenção importante. Note que os atributos começam com um sublinhado: `_nome`, `_matricula`. Trata-se de uma convenção da comunidade Python: o `_` sinaliza que o atributo é **interno** à classe e não deve ser acessado diretamente de fora.

O Python não impede o acesso (é possível ler `a1._nome`), mas a prática é desencorajada. O caminho correto é usar os métodos públicos (`get_nome`), que fazem parte do **contrato** da classe; o que está marcado com `_` pode ser alterado pelo autor da classe a qualquer momento.

```python
a1._nome
```

```python
a1.get_nome()
```

## Fechando o ciclo

Iniciamos a aula com o módulo `circulo.py`, em que tínhamos **funções soltas** recebendo um raio. Podemos agora observar a mesma ideia sob a lente da orientação a objetos: e se o círculo fosse um **objeto**, guardando sua posição e seu raio, e capaz de desenhar a si mesmo?

```python
import math

class Circulo:
    def __init__(self, x, y, raio):
        self.x = x
        self.y = y
        self.raio = raio
    def area(self):
        return math.pi * self.raio**2
    def desenhar(self):
        plt.gca().add_patch(plt.Circle((self.x, self.y), self.raio, alpha=0.4))
```

```python
c = Circulo(0, 0, 2)
c.area()
```

Cada círculo carrega os seus próprios dados (`x`, `y`, `raio`) e o seu comportamento (`area`, `desenhar`). Vamos criar vários e pedir que cada um se desenhe.

```python
circulos = [Circulo(0, 0, 2), Circulo(3, 1, 1), Circulo(-1, 2, 1.5)]

for c in circulos:
    c.desenhar()

plt.xlim(-4, 6)
plt.ylim(-3, 5)
plt.gca().set_aspect('equal')
plt.show()
```

Cada objeto sabe desenhar a si mesmo na posição correta, com o tamanho correto. É a essência da orientação a objetos: dados e comportamento reunidos em um único pacote.

## Recapitulando

Esta aula tratou de **organizar** código que cresceu:

- **Módulos** são arquivos `.py` reaproveitados via `import`. A notação de ponto (`modulo.nome`) mantém os espaços de nomes separados; `import ... as ...` permite renomear o módulo localmente.
- **Arquivos** persistem dados entre execuções. Abra-os com `open`, prefira o bloco `with` (que fecha automaticamente) e escolha o modo: `'r'` lê, `'w'` reescreve, `'a'` acrescenta.
- **Exceções** tratam situações inesperadas. `try`/`except` lida com o erro sem interromper o programa; `raise` levanta uma exceção própria; `assert` verifica condições e falha cedo.
- **Classes** definem tipos novos que reúnem **dados** (atributos) e **comportamento** (métodos). O `__init__` constrói o objeto, o `self` referencia o próprio objeto, os **dunder methods** (`__str__`, `__add__`, `__lt__`) integram o tipo à linguagem, e a **herança** (com `super` e `isinstance`) permite reaproveitar e especializar.
