```python
# @title 💡 Solução C2 (clique em "Mostrar código" para revelar)
# Solução C2
def min_max(lista):
    menor = maior = lista[0]
    for x in lista[1:]:
        if x < menor:
            menor = x
        if x > maior:
            maior = x
    return (menor, maior)

assert min_max([3, 1, 4, 1, 5, 9, 2, 6]) == (1, 9)
assert min_max([7]) == (7, 7)
assert min_max([-5, -3, -10, -1]) == (-10, -1)
print("OK!")
```

**Casos de teste:**

| `s` | resultado |
|---|---|

```python
# @title 💡 Solução C3 (clique em "Mostrar código" para revelar)
# Solução C3
def frequencia(s):
    contagem = {}
    for c in s:
        if c in contagem:
            contagem[c] = contagem[c] + 1
        else:
            contagem[c] = 1
    return contagem

assert frequencia("banana") == {"b": 1, "a": 3, "n": 2}
assert frequencia("aaaa") == {"a": 4}
assert frequencia("") == {}
print("OK!")
```

```python
# @title 💡 Solução D2 (clique em "Mostrar código" para revelar)
# Solução D2
def potencia(base, expoente):
    if expoente == 0:
        return 1
    return base * potencia(base, expoente - 1)

assert potencia(2, 10) == 1024
assert potencia(5, 0) == 1
assert potencia(3, 4) == 81
assert potencia(7, 1) == 7
print("OK!")
```
