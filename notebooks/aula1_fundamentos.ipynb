## O que é computação?

Tem dois jeitos de saber alguma coisa. Um deles é o **conhecimento declarativo**: uma afirmação sobre a verdade. Por exemplo, *a raiz quadrada de x é um número y tal que y vezes y é igual a x*. Isso é útil para **reconhecer** uma resposta certa, mas não diz como **chegar** nela.

O outro jeito é o **conhecimento imperativo**: uma receita. *Pegue um chute inicial, melhore o chute, repita até estar bom o suficiente*. Isso é um **algoritmo**.

Um programa de computador é simplesmente um algoritmo escrito numa linguagem que a máquina entende. Quando a gente escreve Python, está descrevendo passo a passo o que a máquina deve fazer para transformar uma entrada numa saída.

## Variáveis: rótulos, não caixas

Até agora a gente calculou coisas e jogou fora. Para guardar um valor e usar depois, usamos uma **variável**.

A intuição clássica é pensar numa variável como uma *caixinha* onde a gente guarda um valor. Essa imagem funciona, mas tem uma melhor: pense na variável como um **rótulo** colado no valor. O valor existe na memória por si só; o nome é só uma etiqueta para se referir a ele.

```python
raio = 5
pi = 3.14159
area = pi * raio ** 2
area
```

```python
raio = 10
area = pi * raio ** 2
area
```

Quando a gente faz `raio = 10`, o nome `raio` *desliga* do valor `5` e *gruda* no valor `10`. O `5` continua existindo (em outro contexto, ele poderia ser coletado pelo *garbage collector*), mas o rótulo mudou de dono.

Algumas convenções de nomes que vamos seguir o curso inteiro:

- Letras minúsculas, com `snake_case` para separar palavras: `numero_de_amostras`.
- Nomes em inglês são preferíveis em código de produção, mas para didática a gente mistura sem culpa.
- Nomes devem **descrever o que a variável é**, não como ela é usada. Prefira `temperatura` a `t`.

## Tomada de decisão: branching

Até aqui o programa era uma linha reta. Agora vamos ensinar ele a **bifurcar**: fazer uma coisa se uma condição for verdadeira, outra coisa se não for. Isso é o `if`.

```python
idade = 20

if idade >= 18:
    print('Maior de idade')
```

Repare em duas coisas. Primeiro, o `:` no final da linha. Segundo, a **indentação** (esses quatro espaços antes do `print`). Em Python, indentação não é estética — é sintaxe. É como a linguagem sabe o que está *dentro* do `if` e o que está fora.

```python
idade = 15

if idade >= 18:
    print('Maior de idade')
else:
    print('Menor de idade')
```

Quando temos mais de duas opções, encadeamos com `elif`. Um exemplo clássico: classificar uma nota.

```python
nota = 7.5

if nota >= 9:
    conceito = 'A'
elif nota >= 7:
    conceito = 'B'
elif nota >= 5:
    conceito = 'C'
else:
    conceito = 'D'

conceito
```

Detalhe importante: o Python avalia as condições **de cima para baixo** e para na primeira que for verdadeira. Por isso a ordem das comparações importa.

## Repetir tarefas: iteração

A outra superpotência do programa é **repetir**. Tem dois jeitos principais de repetir em Python: `while` (enquanto uma condição for verdadeira) e `for` (para cada item de uma sequência).

```python
contador = 0

while contador < 5:
    print(contador)
    contador = contador + 1
```

O `while` é flexível, mas perigoso: se a condição nunca virar falsa, o programa fica preso num **loop infinito**. Sempre se pergunte: *o que dentro do laço vai eventualmente tornar a condição falsa?* Aqui é a linha `contador = contador + 1`.

```python
for i in range(5):
    print(i)
```

O `for` é mais seguro porque o número de iterações é definido pela sequência sobre a qual estamos iterando. Aqui `range(5)` gera os números `0, 1, 2, 3, 4`.

Vamos calcular a soma de `1` até `100` usando um `for`.

```python
soma = 0
for n in range(1, 101):
    soma = soma + n

soma
```

Gauss, quando era criança, descobriu uma fórmula fechada para esse mesmo problema: $\frac{n(n+1)}{2}$. Para $n = 100$, dá `5050`. Boa hora para a primeira lição filosófica do curso: **um problema pode ter vários algoritmos**, e alguns são muito mais eficientes que outros. O `for` resolve em 100 passos; a fórmula resolve em 1.

```python
n = 100
n * (n + 1) // 2
```

Iterar não é exclusividade dos números. Tudo que for uma sequência pode ser iterado, inclusive strings.

```python
for letra in 'Python':
    print(letra)
```

## Pensar como cientista da computação

Agora a gente junta tudo num primeiro problema interessante: **achar a raiz cúbica de um número inteiro**, sem usar `x ** (1/3)`. Por que? Porque o caminho até a resposta vai mostrar dois estilos de pensar bem diferentes.

**Primeira estratégia: enumeração exaustiva.**

A ideia é boba e poderosa: comece em zero e teste, um a um, se aquele número ao cubo é o que a gente procura. Pare quando achar ou passar do alvo.

```python
x = 27
chute = 0

while chute ** 3 < x:
    chute = chute + 1

if chute ** 3 == x:
    print(f'Raiz cúbica de {x} é {chute}')
else:
    print(f'{x} não é um cubo perfeito')
```

```python
x = 1000000
chute = 0
passos = 0

while chute ** 3 < x:
    chute = chute + 1
    passos = passos + 1

print('Resposta:', chute, 'em', passos, 'passos')
```

Funciona. Mas vocês já viram o problema: para achar a raiz cúbica de um milhão, foram cem passos. Para um bilhão, seriam mil. O número de passos cresce *proporcionalmente* ao tamanho da resposta. Isso se chama **crescimento linear**, e em geral é um sinal de que dá para fazer melhor.

**Segunda estratégia: busca por bisseção.**

A intuição vem do jogo do *maior ou menor*. Se eu peço para vocês adivinharem um número entre 1 e 100, e a cada chute eu digo se vocês passaram ou ficaram aquém, qual é a estratégia ótima? Chutar **a metade do intervalo restante** toda vez. A cada chute, vocês descartam metade do espaço de busca.

Vamos aplicar isso para achar uma raiz quadrada aproximada (a cúbica é a mesma ideia).

```python
x = 1000000
epsilon = 0.01
baixo = 0
alto = x
chute = (baixo + alto) / 2
passos = 0

while abs(chute ** 2 - x) > epsilon:
    if chute ** 2 < x:
        baixo = chute
    else:
        alto = chute
    chute = (baixo + alto) / 2
    passos = passos + 1

print('Raiz aproximada:', chute, 'em', passos, 'passos')
```

A enumeração exaustiva pediu mil passos para a raiz quadrada de um milhão. A bisseção resolveu o mesmo problema em algumas dezenas de passos. Esse comportamento — dobrar o tamanho do problema e só adicionar **um passo** a mais — chama-se **crescimento logarítmico**, e a gente vai voltar a esse conceito com calma na aula 4.

Vamos visualizar a diferença plotando, para vários tamanhos de `x`, quantos passos cada estratégia precisa.

```python
def passos_enumeracao(x):
    chute = 0
    while chute ** 2 < x:
        chute += 1
    return chute

def passos_bisseccao(x, epsilon=0.01):
    baixo, alto = 0, max(x, 1)
    chute = (baixo + alto) / 2
    passos = 0
    while abs(chute ** 2 - x) > epsilon:
        if chute ** 2 < x:
            baixo = chute
        else:
            alto = chute
        chute = (baixo + alto) / 2
        passos += 1
    return passos
```

```python
tamanhos = [10, 100, 1000, 10000, 100000, 1000000]
enum = [passos_enumeracao(x) for x in tamanhos]
bis = [passos_bisseccao(x) for x in tamanhos]

plt.plot(tamanhos, enum, marker='o', label='Enumeração')
plt.plot(tamanhos, bis, marker='o', label='Bisseção')
plt.xscale('log')
plt.xlabel('Tamanho de x')
plt.ylabel('Passos')
plt.legend()
plt.show()
```

O gráfico fala por si. A enumeração explode; a bisseção mal sai do chão. **A escolha do algoritmo é mais determinante para a velocidade do programa do que qualquer otimização de baixo nível.**
