```python
# @title 💡 Solução D2 (clique em "Mostrar código" para revelar)
class ContaBancaria:
    def __init__(self, saldo_inicial=0):
        self._saldo = saldo_inicial
    def depositar(self, valor):
        self._saldo += valor
    def sacar(self, valor):
        if valor > self._saldo:
            raise ValueError('saldo insuficiente')
        self._saldo -= valor
    def get_saldo(self):
        return self._saldo

c = ContaBancaria(100)
c.depositar(50)
assert c.get_saldo() == 150
c.sacar(30)
assert c.get_saldo() == 120
try:
    c.sacar(1000)
    levantou = False
except ValueError:
    levantou = True
assert levantou
assert c.get_saldo() == 120
print('OK!')
```

```python
# @title 💡 Solução D3 (clique em "Mostrar código" para revelar)
class Ponto:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return '(' + str(self.x) + ', ' + str(self.y) + ')'
    def __eq__(self, outro):
        return self.x == outro.x and self.y == outro.y
    def __add__(self, outro):
        return Ponto(self.x + outro.x, self.y + outro.y)

assert str(Ponto(1, 2)) == '(1, 2)'
assert Ponto(1, 2) == Ponto(1, 2)
assert not (Ponto(1, 2) == Ponto(3, 4))
assert (Ponto(1, 2) + Ponto(3, 4)) == Ponto(4, 6)
print('OK!')
```

| objeto | `salario()` |
|---|---|

Além disso, um `Gerente` deve continuar sendo um `Funcionario`, ou seja, `isinstance(gerente, Funcionario)` é `True`.

```python
class Funcionario:
    def __init__(self, nome, salario_base):
        # TODO: guardar nome e salario_base
        ...
    def salario(self):
        # TODO: devolver o salario base
        ...

class Gerente(Funcionario):
    def __init__(self, nome, salario_base, bonus):
        # TODO: usar super().__init__ e guardar o bonus
        ...
    def salario(self):
        # TODO: salario base + bonus
        ...

print(Gerente('Beto', 3000, 1000).salario())
```

```python
# @title 💡 Solução D4 (clique em "Mostrar código" para revelar)
class Funcionario:
    def __init__(self, nome, salario_base):
        self.nome = nome
        self.salario_base = salario_base
    def salario(self):
        return self.salario_base

class Gerente(Funcionario):
    def __init__(self, nome, salario_base, bonus):
        super().__init__(nome, salario_base)
        self.bonus = bonus
    def salario(self):
        return self.salario_base + self.bonus

assert Funcionario('Ana', 3000).salario() == 3000
g = Gerente('Beto', 3000, 1000)
assert g.salario() == 4000
assert isinstance(g, Funcionario)
assert g.nome == 'Beto'
print('OK!')
```
